## 1. Imports

Um zu plotten müssen wir das entsprechende Modul, in dem die notwendigen Funktionen vorhanden sind, importieren. Führen Sie hieryu den folgenden Code aus.

```python
import matplotlib.pyplot as plt
```

Wir werden Scatter- und Linienplots basierend auf Punkten plotten. Hierzu brauchen wir eine Liste der $x$- und eine Liste der $y$-Werte. Beide Listen müssen gleich lang sein. 

Um es etwas einfacher zu machen, existiert eine Funktion, die eine Liste von Punkten in zwei Listen, eine mit $x$- und eine mit $y$-Werten, umwandelt.

Führen Sie den folgenden Code aus, um die Funktion zu importieren.

```python
from data.plots import punkteInXUndY
```

In der folgenden Zelle sehen Sie, wie die Funktion genutzt wird:

```python
# ein paar Punkte (x,y) in einer Liste [...]
punkteListe=[(1,1),(1,2),(3,4),(5,3)]

# hier rufen wir die Funktion auf und erzeugen die Listen x und y
x,y=punkteInXUndY(punkteListe)

# Ausgabe von x
print(x)

# Ausgabe von y
print(y)
```

## 2. Ein erster Plot

Mit der Funktion `plt.plot()` wird ein Linienplot erstellt. Die Funktion erhält die Listen `x` und `y` der $x$/ und $y$-Werte.

```python
# der erste Plot

# die Punkte
punkteListe=[(1,1),(1.5,0.5),(2,4),(4,2)]

# erzeuge die Listen x und y
x,y=punkteInXUndY(punkteListe)

# plotten
plt.plot(x,y)
```

Die Punkte werden in der angegebenen Reihenfolge verarbeitet. Eine gänderte Reihenfolge der gleichen Punkte ergibt:

```python
# geänderte Reihenfolge

# die Punkte
punkteListe=[(2,4),(1,1),(1.5,0.5),(4,2)]

# erzeuge die Listen x und y
x,y=punkteInXUndY(punkteListe)

# plotten
plt.plot(x,y)
```

## 3. Formatierung

Man kann den Plot formatieren, indem man nach dem `y` z.B. eine der folgenden Zeichenketten angibt:

### Punkte

|Zeichenkette     | Marker           |
|:----:           |:-                |
|`"."`              |kleine Punkte     |
|`"o"`              |größere Punkte    |
|`"*"`              |Sterne            |
|`"^"`              |Dreiecke          |



Testen Sie das ruhig aus:

```python
# Code zum Testen

# die Punkte
punkteListe=[(1,1),(2,2),(3,2.5),(4,4)]

# erzeuge die Listen x und y
x,y=punkteInXUndY(punkteListe)

# plotten
plt.plot(x,y,'o')
    
```

### Linien


|Zeichenkette     | Linie           |
|:----:           |:-               |
|`"-"`            |durchgezogen    |
|`"--"`           |gestrichelt      |
|`":"`            |gepunktet        |


Testen Sie das ruhig aus:

```python
# Code zum Testen

# die Punkte
punkteListe=[(1,1),(2,2),(3,2.5),(4,4)]

# erzeuge die Listen x und y
x,y=punkteInXUndY(punkteListe)

# plotten
plt.plot(x,y,'--')
    
```

### Linien UND Marker

Man kann jeweils eine Formatierungsangabe für die Linien und den Marker beliebig kombinieren. Achten Sie darauf, dass sich **eine** neue Zeichenkette ergibt.

Zum Beispiel:

|Zeichenkette     | Linie                                           |
|:----:           |:-                                               |
|`"-."`           | kleine Punkte mit durchgezogener Linie verbunden|
|`".-"`           | kleine Punkte mit durchgezogener Linie verbunden|
|`"^:"`           | Dreiecke mit gestrichelter Linie verbunden      |

etc.


Testen Sie das ruhig aus:

```python
# Code zum Testen

# die Punkte
punkteListe=[(1,1),(2,2),(3,2.5),(4,4)]

# erzeuge die Listen x und y
x,y=punkteInXUndY(punkteListe)

# plotten
plt.plot(x,y,'^:')
    
```

## 4. Beschriftungen

Es gibt die Möglichkeit, der Abbildung Beschriftungen hinzuzufügen. Hierzu gehören:

- `plt.title("text")`: fügt `text` als Titel der Abbildung hinzu
- `plt.xlabel("xText")`: beschriftet die $x$-Achse mit `xText`
- `plt.ylabel("yText")`: beschriftet die $y$-Achse mit `yText`

**Beispiel:**

```python
# Beispiel

# die Punkte
punkteListe=[(1,1),(2,2),(3,2.5),(4,4)]

# erzeuge die Listen x und y
x,y=punkteInXUndY(punkteListe)

# plotten
plt.plot(x,y,'^:')

plt.title("Titel der Abbildung")
plt.xlabel("Beschriftung der x-Achse")
plt.ylabel("Beschriftung der y-Achse")
```
